```python
import yfinance as yf
from datetime import datetime, timedelta
symbol = "QQQ"
interval = "15m"

max_dhm = {
    "1m" : (6,23,59),
    "5m" : (59, 23, 55),
    "15m" : (59, 23, 45),
    "30m" : (59, 23, 30),
    "1h" : (729, 23, 0),
    "1d" : (730, 0, 0)}
for interval in max_dhm.keys():
    current_date = datetime.now() + timedelta(hours = 3)
    start_date = current_date - timedelta(days = max_dhm[interval][0], hours=max_dhm[interval][1], minutes=max_dhm[interval][2])
    data = yf.download(symbol, start=start_date, interval=interval)  # load data of symbol
    data.to_csv(f"./data/{symbol}_{interval}.csv", index=True)
```
